```python
# Giả sử df_check là DataFrame chứa dữ liệu của df_2
df_check = pd.DataFrame(df_2)

# Hàm kiểm tra ngày hợp lệ
def is_valid_date(date_str):
    if pd.isna(date_str):  # Kiểm tra nếu là NaN (giá trị thiếu)
        return False
    try:
        # Sử dụng đúng datetime.strptime
        return datetime.strptime(date_str, '%Y-%m-%d') is not None
    except ValueError:
        return False

# Hàm kiểm tra logic giữa ngày sản xuất và hạn sử dụng
def check_dates(df_check):
    invalid_nsx = []  # Danh sách các ngày sản xuất không hợp lệ
    invalid_hsd = []  # Danh sách các hạn sử dụng không hợp lệ
    invalid_logic = []  # Danh sách các bản ghi có logic sai

    for i, row in df_check.iterrows():
        nsx = str(row['NSX'])  # Chuyển đổi thành chuỗi
        hsd = str(row['HSD'])  # Chuyển đổi thành chuỗi
        
        # Kiểm tra ngày sản xuất có hợp lệ không
        if not is_valid_date(nsx):
            invalid_nsx.append(nsx)

        # Kiểm tra hạn sử dụng có hợp lệ không
        if not is_valid_date(hsd):
            invalid_hsd.append(hsd)

        # Nếu cả hai ngày đều hợp lệ, kiểm tra logic ngày
        if is_valid_date(nsx) and is_valid_date(hsd):
            nsx_date = datetime.strptime(nsx, '%Y-%m-%d')
            hsd_date = datetime.strptime(hsd, '%Y-%m-%d')
            if nsx_date >= hsd_date:  # Thêm điều kiện "bằng nhau" không hợp lệ
                invalid_logic.append((nsx, hsd))
    
    return invalid_nsx, invalid_hsd, invalid_logic

# Gọi hàm để kiểm tra dữ liệu
invalid_nsx, invalid_hsd, invalid_logic = check_dates(df_check)

# In kết quả
print("Ngày sản xuất không hợp lệ:", invalid_nsx)
print("Hạn sử dụng không hợp lệ:", invalid_hsd)
print("Ngày sản xuất vượt qua hoặc bằng hạn sử dụng:", invalid_logic)
```

```python
# Tạo cột 'Age group' dựa trên cột 'Age'
def categorize_age(age):
    if age >= 18 and age <= 30:
        return 'Thiếu niên'
    elif age >= 31 and age <= 60:
        return 'Trung niên'
    elif age > 60:
        return 'Cao tuổi'
    else:
        return 'Dưới 18 tuổi'

# Áp dụng hàm trên cho cột Age để tạo ra cột Age Group
df_3['Age group'] = df_3['Age'].apply(categorize_age)
```

```python
# Sau khi check thấy các giá trị trong 2 feature Looking at item (s), Holding at item (s) có các giá trị bằng không
# Không holding nhưng pick sản phẩm để vào bag thì hợp lý nhưng không nhìn sản phẩm dù chỉ 0.1 s 
# nhưng lại có những thao tác mua hàng tiếp theo thì không hợp lý, nên phương Phá giải quyết là
# các row Looking at item = 0 nhưng lại có giá trị Holding the item (s) khác 0 thì đưa giá trị Holding the item (s) về 0
# các row Looking at item = 0 thì Fill các dòng dữ liệu Picking up item, Returning item, Putting item into bag, 
#Taking item out of bag, Putting item into bag in the 2nd timebằng giá trị False

# Cập nhật giá trị 'Holding the item (s)' về 0 khi 'Looking at item (s)' = 0 và 'Holding the item (s)' khác 0
df_3.loc[(df_3['Looking at item (s)'] == 0) & (df_3['Holding the item (s)'] != 0), 'Holding the item (s)'] = 0

# Cập nhật các cột 'Picking up item', 'Returning item', 'Putting item into bag', 'Taking item out of bag', 'Putting item into bag in the 2nd time'
# thành False nếu 'Looking at item (s)' = 0
columns_to_update = ['Picking up item', 'Returning item', 'Putting item into bag', 'Taking item out of bag', 'Putting item into bag in the 2nd time']
df_3.loc[df_3['Looking at item (s)'] == 0, columns_to_update] = False
```

```python
# Sau khi check info thấy được 'Putting item into bag','Taking item out of bag','Putting item into bag in the 2nd time' có chứa
# những giá trị null, dưới đây là các trường hợp dùng để fill null cũng như là giải quyết bằng cách khác


# Điều kiện 1: 'Putting item into bag' null, 'Taking item out of bag' null, 'Putting item into bag in the 2nd time' null -> fill False
df_3.loc[df_3['Putting item into bag'].isnull() & df_3['Taking item out of bag'].isnull() & df_3['Putting item into bag in the 2nd time'].isnull(), 
         ['Putting item into bag', 'Taking item out of bag', 'Putting item into bag in the 2nd time']] = False

# Điều kiện 2: 'Putting item into bag' null, 'Taking item out of bag' null, 'Putting item into bag in the 2nd time' True -> fill 'Putting item into bag' and 'Taking item out of bag' with True
df_3.loc[df_3['Putting item into bag'].isnull() & df_3['Taking item out of bag'].isnull() & (df_3['Putting item into bag in the 2nd time'] == True), 
         ['Putting item into bag', 'Taking item out of bag']] = True

# Điều kiện 3: 'Putting item into bag' null, 'Taking item out of bag' null, 'Putting item into bag in the 2nd time' False -> fill False
df_3.loc[df_3['Putting item into bag'].isnull() & df_3['Taking item out of bag'].isnull() & (df_3['Putting item into bag in the 2nd time'] == False), 
         ['Putting item into bag', 'Taking item out of bag', 'Putting item into bag in the 2nd time']] = False

# Điều kiện 4: 'Putting item into bag' null, 'Taking item out of bag' True, 'Putting item into bag in the 2nd time' True or False -> fill 'Putting item into bag' with True
df_3.loc[df_3['Putting item into bag'].isnull() & (df_3['Taking item out of bag'] == True) & df_3['Putting item into bag in the 2nd time'].notnull(), 
         'Putting item into bag'] = True

# Điều kiện 5: 'Putting item into bag' null, 'Taking item out of bag' True, 'Putting item into bag in the 2nd time' null -> fill False
df_3.loc[df_3['Putting item into bag'].isnull() & (df_3['Taking item out of bag'] == True) & df_3['Putting item into bag in the 2nd time'].isnull(), 
         ['Putting item into bag', 'Taking item out of bag', 'Putting item into bag in the 2nd time']] = False

# Điều kiện 6: 'Putting item into bag' null, 'Taking item out of bag' False, 'Putting item into bag in the 2nd time' False or True -> fill False
df_3.loc[df_3['Putting item into bag'].isnull() & (df_3['Taking item out of bag'] == False), 
         ['Putting item into bag', 'Taking item out of bag', 'Putting item into bag in the 2nd time']] = False

# Điều kiện 7: 'Putting item into bag' True, 'Taking item out of bag' null, 'Putting item into bag in the 2nd time' null -> fill False
df_3.loc[(df_3['Putting item into bag'] == True) & df_3['Taking item out of bag'].isnull() & df_3['Putting item into bag in the 2nd time'].isnull(), 
         ['Taking item out of bag', 'Putting item into bag in the 2nd time']] = False


# Điều kiện 8: 'Putting item into bag' True, 'Taking item out of bag' True, 'Putting item into bag in the 2nd time' null -> fill 'Putting item into bag in the 2nd time' with False
df_3.loc[(df_3['Putting item into bag'] == True) & 
         (df_3['Taking item out of bag'] == True) & 
         df_3['Putting item into bag in the 2nd time'].isnull(), 
         'Putting item into bag in the 2nd time'] = False

# Điều kiện 9: 'Putting item into bag' True, 'Taking item out of bag' False, 'Putting item into bag in the 2nd time' null -> fill 'Putting item into bag in the 2nd time' with False
df_3.loc[(df_3['Putting item into bag'] == True) & 
         (df_3['Taking item out of bag'] == False) & 
         df_3['Putting item into bag in the 2nd time'].isnull(), 
         'Putting item into bag in the 2nd time'] = False

# Điều kiện 10: 'Putting item into bag' True, 'Taking item out of bag' null, 'Putting item into bag in the 2nd time' True -> fill 'Taking item out of bag' with True
df_3.loc[(df_3['Putting item into bag'] == True) & 
         df_3['Taking item out of bag'].isnull() & 
         (df_3['Putting item into bag in the 2nd time'] == True), 
         'Taking item out of bag'] = True

# Điều kiện 11: 'Putting item into bag' True, 'Taking item out of bag' null, 'Putting item into bag in the 2nd time' False -> fill 'Taking item out of bag' with False
df_3.loc[(df_3['Putting item into bag'] == True) & 
         df_3['Taking item out of bag'].isnull() & 
         (df_3['Putting item into bag in the 2nd time'] == False), 
         'Taking item out of bag'] = False

# Điều kiện 12: 'Putting item into bag' False, 'Taking item out of bag' null, 'Putting item into bag in the 2nd time' null -> fill both 'Taking item out of bag' and 'Putting item into bag in the 2nd time' with False
df_3.loc[(df_3['Putting item into bag'] == False) & 
         df_3['Taking item out of bag'].isnull() & 
         df_3['Putting item into bag in the 2nd time'].isnull(), 
         ['Taking item out of bag', 'Putting item into bag in the 2nd time']] = False

# Điều kiện 13: 'Putting item into bag' False, 'Taking item out of bag' null, 'Putting item into bag in the 2nd time' False -> fill 'Taking item out of bag' with False
df_3.loc[(df_3['Putting item into bag'] == False) & 
         df_3['Taking item out of bag'].isnull() & 
         (df_3['Putting item into bag in the 2nd time'] == False), 
         'Taking item out of bag'] = False

# Điều kiện 14: 'Putting item into bag' False, 'Taking item out of bag' False, 'Putting item into bag in the 2nd time' null -> fill 'Putting item into bag in the 2nd time' with False
df_3.loc[(df_3['Putting item into bag'] == False) & 
         (df_3['Taking item out of bag'] == False) & 
         df_3['Putting item into bag in the 2nd time'].isnull(), 
         'Putting item into bag in the 2nd time'] = False
```

```python
# Biến những hành vi mua sắm không hợp lý thành Nan, vừa để lọc dữ liệu cho những phân tích cần đến phía sau, vừa không làm mất
#đi dữ liệu trong row dùng cho những phân tích khác

# Điều kiện 1: "Holding the bag" là False nhưng có ít nhất một trong ba feature dưới đây là True:
# "Putting item into bag", "Taking item out of bag", "Putting item into bag in the 2nd time"
condition_1 = (df_3['Holding the bag'] == False) & (
    (df_3['Putting item into bag'] == True) |
    (df_3['Taking item out of bag'] == True) |
    (df_3['Putting item into bag in the 2nd time'] == True)
)

# Điều kiện 2: "Picking up item" là True nhưng không có "Putting item into bag" và "Returning item"
condition_2 = (df_3['Picking up item'] == True) & ~(
    (df_3['Putting item into bag'] == True) |
    (df_3['Returning item'] == True)
)
 
# Điều kiện 3: "Returning item" xảy ra mà không có "Picking up item" 
condition_3 = (df_3['Returning item'] == True) & (
    (df_3['Picking up item'] != True) 
)

# Điều kiện 4: "Putting item into bag in the 2nd time" xảy ra mà không có đồng thời cả 3 hành động:
# "Picking up item", "Putting item into bag", "Taking item out of bag"
condition_4 = (df_3['Putting item into bag in the 2nd time'] == True) & ~(
    (df_3['Picking up item'] == True) &
    (df_3['Putting item into bag'] == True) &
    (df_3['Taking item out of bag'] == True)
)

# Điều kiện 5: "Returning item" xảy ra mà có 'Putting item into bag':
condition_5 = (df_3['Returning item'] == True) & (df_3['Putting item into bag'] == True)

# Điều kiện 6: "Taking item out of bag" xảy ra mà không có 'Putting item into bag':
condition_6 = (df_3['Taking item out of bag'] == True) & (df_3['Putting item into bag'] != True)

# Kết hợp tất cả các điều kiện
all_conditions = condition_1 | condition_2 | condition_3 | condition_4 | condition_5 | condition_6

df_3.loc[all_conditions, ['Holding the bag', 
                          'Picking up item', 
                          'Returning item', 
                          'Putting item into bag', 
                          'Taking item out of bag', 
                          'Putting item into bag in the 2nd time']] = np.nan

df_3
```

```python
#1.Thống kê 5 mặt hàng có tổng thời gian nhìn và cầm xem lâu nhất?

# Tạo một cột tổng thời gian nhìn và cầm xem
df_3c['Total_time'] = df_3c['Looking at item (s)'] + df_3c['Holding the item (s)']

# Nhóm dữ liệu theo Shelf ID và Item ID, rồi tính tổng thời gian cho mỗi nhóm
grouped_df = df_3c.groupby(['Shelf ID', 'Item ID'], as_index=False)['Total_time'].sum()

# Sắp xếp theo tổng thời gian từ cao xuống thấp
sorted_df = grouped_df.sort_values(by='Total_time', ascending=False)

# Lấy 5 mặt hàng có tổng thời gian lâu nhất
top_5_items = sorted_df.head(5)

# Hiển thị kết quả
print(top_5_items)
```

```python
#2.Thống kê 5 mặt hàng thường được cầm lên rồi trả lại nhiều nhất?

# Đếm số lần "Returning item" là 1 cho mỗi kết hợp của 'Shelf ID' và 'Item ID'
top_items_returned = df_3c[df_3c['Returning item'] == 1][['Shelf ID', 'Item ID']].value_counts().head(5)

print("Top 5 mặt hàng thường được cầm lên rồi trả lại nhiều nhất:")
print(top_items_returned)
```

```python
#3.Các nhóm khách hàng theo độ tuổi (Thiếu niên: 18 - 30; Trung niên: 31 - 60; Cao tuổi: > 60) mua mặt hàng nào nhiều nhất? 

# Lọc các hàng có "Be purchased" là True
purchased_items = df_3c[df_3c['Be purchased'] == True]

# Đếm số lần mua cho mỗi nhóm tuổi, Shelf ID và Item ID
top_items_by_age_group = purchased_items.groupby(['Age group', 'Shelf ID', 'Item ID']).size().reset_index(name='Count')

# Tìm mặt hàng được mua nhiều nhất cho mỗi nhóm tuổi
most_purchased_items = top_items_by_age_group.loc[top_items_by_age_group.groupby('Age group')['Count'].idxmax()]

print("Mặt hàng được mua nhiều nhất theo nhóm tuổi và kệ hàng:")
print(most_purchased_items)
```

```python
#4. Ngày nào trong tuần có doanh thu cao nhất?

# Lọc các sản phẩm đã được mua (Be purchased = True)
purchased_items = df_3c[df_3c['Be purchased'] == True]

# Kết hợp df_3c với df_2c dựa trên 'Shelf ID' và 'Item ID' để lấy thông tin giá cả
merged_data = pd.merge(purchased_items, df_2c[['Shelf ID', 'Item ID', 'Price (vnđ)']], on=['Shelf ID', 'Item ID'], how='left')

# Tính tổng doanh thu cho mỗi ngày trong tuần
revenue_by_day = merged_data.groupby('Day of the week')['Price (vnđ)'].sum().reset_index(name='Total Revenue')

# Sắp xếp để tìm ngày có doanh thu cao nhất
highest_revenue_day = revenue_by_day.sort_values(by='Total Revenue', ascending=False).head(1)

print("Ngày trong tuần có doanh thu cao nhất:")
highest_revenue_day
```

```python
#5.Trong 3 nhóm tuổi sau: Thiếu niên (18 - 30),  Trung niên (31 - 60), Cao tuổi: (> 60), 
#nhóm tuổi nào có số người đi siêu thị nhiều nhất?

# Đếm số lượng người đi siêu thị theo nhóm tuổi
people_count_by_age_group = df_3c.groupby('Age group')['Person ID'].nunique().reset_index(name='Number of People')

# Sắp xếp để tìm nhóm tuổi có số người đi siêu thị nhiều nhất
most_frequent_age_group = people_count_by_age_group.sort_values(by='Number of People', ascending=False).head(3)

print("Nhóm tuổi có số người đi siêu thị nhiều nhất:")
print(most_frequent_age_group)
```

```python
#6.Top 5 các mặt hàng giảm giá được người dùng mua nhiều nhất?

# Lọc các mặt hàng giảm giá từ bảng df_2c
discounted_items = df_2c[df_2c['Discount (%)'] > 0]

# Kết hợp df_3c với df_2c dựa trên 'Shelf ID' và 'Item ID' để lấy thông tin giảm giá
merged_data = pd.merge(df_3c, discounted_items[['Shelf ID', 'Item ID']], on=['Shelf ID', 'Item ID'], how='inner')

# Lọc các mặt hàng giảm giá đã được mua (Be purchased = True)
purchased_discounted_items = merged_data[merged_data['Be purchased'] == True]

# Đếm số lượng mua cho mỗi mặt hàng giảm giá
top_discounted_items = purchased_discounted_items.groupby(['Shelf ID', 'Item ID']).size().reset_index(name='Number of Purchases')

# Sắp xếp để tìm top 5 mặt hàng giảm giá được mua nhiều nhất
top_5_discounted_items = top_discounted_items.sort_values(by='Number of Purchases', ascending=False).head(5)

print("Top 5 mặt hàng giảm giá được người dùng mua nhiều nhất:")
print(top_5_discounted_items)
```

```python
#7.Top 5 các mặt hàng được chạy quảng cáo được người dùng mua nhiều nhất?

# Lọc các mặt hàng được chạy quảng cáo trong df_2c
ads_items = df_2c[df_2c['Marketing strategy'] == True]

# Lọc các mặt hàng đã được mua trong df_3c
purchased_items = df_3c[df_3c['Be purchased'] == True]

# Kết hợp df_2c với df_3c dựa trên 'Shelf ID' và 'Item ID' để lấy thông tin giá cả và số lượng mua
merged_data = pd.merge(ads_items, purchased_items, on=['Shelf ID', 'Item ID'], how='inner')

# Tính số lượng mua cho từng mặt hàng
purchase_counts = merged_data.groupby(['Shelf ID', 'Item ID']).size().reset_index(name='Purchase Count')

# Sắp xếp các mặt hàng theo số lượng mua giảm dần và chọn top 5
top_5_items = purchase_counts.sort_values(by='Purchase Count', ascending=False).head(5)

print(top_5_items)

```

```python
#8.Top 3 quầy hàng có thời lượng trung bình quan tâm đến sản phẩm, trên số lượt tương tác,
#là lâu nhất (quan tâm tương ứng với việc nhìn và cầm xem)?

df = pd.DataFrame(df_3c)
df['attention time (s)'] = df['Looking at item (s)'] + df['Holding the item (s)']

# Tính số lượt tương tác (đếm số feature có giá trị True)
df['Interaction count'] = df[['Picking up item', 'Returning item', 'Putting item into bag', 
                              'Taking item out of bag', 'Putting item into bag in the 2nd time']].sum(axis=1)

# Loại bỏ các hàng không có tương tác
df = df[df['Interaction count'] > 0]

# Tính tổng thời gian quan tâm và tổng số tương tác theo từng quầy (Shelf ID)
shelf_group = df.groupby('Shelf ID').agg(
    total_attention_time=('attention time (s)', 'sum'),
    total_interactions=('Interaction count', 'sum')
)

# Tính thời lượng quan tâm trung bình trên mỗi tương tác
shelf_group['Attention time per interaction'] = shelf_group['total_attention_time'] / shelf_group['total_interactions']

# Sắp xếp quầy hàng theo thời lượng trung bình giảm dần và chọn Top 3
top_3_shelves = shelf_group.sort_values(by='Attention time per interaction', ascending=False).head(3)

# Hiển thị kết quả
print(top_3_shelves)

```

```python
#9.Top 3 quầy hàng có số sản phẩm được mua nhiều nhất?

df = pd.DataFrame(df_3c)

# Lọc những sản phẩm đã được mua (Be purchased == True)
df_purchased = df[df['Be purchased'] == True]

# Đếm số lượng sản phẩm được mua theo từng quầy hàng (Shelf ID)
shelf_purchases = df_purchased.groupby('Shelf ID').size().reset_index(name='Products Purchased')

# Sắp xếp theo số lượng sản phẩm được mua giảm dần và chọn Top 3
top_3_shelves = shelf_purchases.sort_values(by='Products Purchased', ascending=False).head(3)

# Hiển thị kết quả
print(top_3_shelves)
```

```python
#10. Người dùng có thói quen di chuyển giữa 2 quầy hàng nào nhiều nhất?

df = pd.DataFrame(df_3c)

# Tạo một cột 'Previous Shelf ID' để theo dõi quầy trước đó
df['Previous Shelf ID'] = df.groupby('Person ID')['Shelf ID'].shift(1)

# Loại bỏ các hàng không có giá trị 'Previous Shelf ID' (tức là lượt di chuyển đầu tiên của mỗi người)
df_moves = df.dropna(subset=['Previous Shelf ID'])

# Chỉ giữ lại các hàng có sự thay đổi quầy (quầy trước đó khác quầy hiện tại)
df_moves = df_moves[df_moves['Previous Shelf ID'] != df_moves['Shelf ID']]

# Tạo cột 'Move' để đại diện cho việc di chuyển từ quầy trước đó sang quầy hiện tại
df_moves['Move'] = list(zip(df_moves['Previous Shelf ID'], df_moves['Shelf ID']))

# Đếm số lần di chuyển giữa các quầy
move_counts = df_moves['Move'].value_counts()

# In ra kết quả
print(move_counts)

```
